```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('student-mat-pass-or-fail.csv')
```

```python
df.head()
```

```python
df1=df.copy()
```

```python
df1=df1.drop(['school','sex','address','famsize','Pstatus','Medu','Fedu','traveltime','nursery','higher','internet','romantic','Dalc','Walc','activities'],axis=1)
```

```python
df1.head()
```

```python
x=df1.drop('pass',axis=1)
y=df1['pass']
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
model=Sequential([
    Flatten(input_shape=(14,)),
    Dense(32,activation='relu'),
    Dense(64,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=30,batch_size=32,validation_data=(xtest,ytest))
```

```python
model.summary()
```

```python
import warnings
```

```python
model.save('ANN1.h5')
```

```python
l=(18,3,3,0,0,0,5,3,3,4,0,13,14,14)
arr=np.array(l).reshape(1,-1)
arr
```

```python
ans=model.predict(arr)
ans
```

```python
if(ans>=0.5):
  print('Pass!')
else:
  print('Fail only')
```
